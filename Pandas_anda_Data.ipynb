<h1><b>Pandas</h1>
<h6><i> Your key to data science world

Pandas is a Python library used for working with data sets. It is a Python package that offers various data structures and operations for manipulating numerical data and time series. It has functions for analyzing, cleaning, exploring, and manipulating data.

<h1><b>Why Use Pandas?

<p>Pandas allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.</p>

Python-Pandas-Applications.png

<h1><b>What Can Pandas Do?


<p>Pandas gives you answers about the data. Like:

*   Is there a correlation between two or more columns?<br>
*   What is average value?<br>
*   Max value?<br>
*   Min value?<br>


Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This is called cleaning the data.</p>

<h4><b>Slicing

````python
# creating a pandas dataframe
df_0 = df[['Min','Max']]
df_0
````

<h4><b>Slicing Using iloc

````python
# Slicing rows in data frame
df1 = df.iloc[0:4]

# data frame after slicing
df1
````

````python

# Slicing columnss in data frame
df1 = df.iloc[1:6, 0:3]

# data frame after slicing
df1
````

````python
df_2 = df.iloc[2, 3]
df_2
````

<h1><b>Statistics

````python
df_mean=df.Max.mean()   #df['Min']mean() we sliced it
print(df_mean)
````

````python
df_median=df.Min.median()
print(df_median)
````

````python
df_quantile=df.Min.quantile(0.25)  #quantile(0.25),quantile(0.50),quantile(0.75)
print(df_quantile)
````
