# **Plotting x and y points**

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.



# **Basic Plot**

```python
import matplotlib.pyplot as plt
import numpy as np
# Create an array of x-values
xpoints = np.array([1, 8])

# Create an array of y-values
ypoints = np.array([3, 10])

# Plot the x-values against the y-values
plt.plot(xpoints, ypoints)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of y-values
ypoints = np.array([3, 8, 1, 10])

# Plot the y-values with circular markers ('o')
plt.plot(ypoints )

# Add a title to the plot
plt.title("Sports Watch Data")

# Label the x-axis
plt.xlabel("Average Pulse")

# Label the y-axis
plt.ylabel("Calorie Burnage")

# Display the plot
plt.show()
```

# **Pie** **Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

# Create an array of values representing the sizes of each slice of the pie chart
y = np.array([35, 25, 25, 15])

# Create a pie chart
plt.pie(y)

# Display the pie chart
plt.show()
```

# **Histogram**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create the histogram
plt.hist(data)

# Add a title and labels
plt.title('Simple Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

## **Heatmap**




```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2D array (matrix) of values
data = np.array([[30, 20, 10],
                 [40, 50, 60],
                 [70, 80, 90]])

# Create the heatmap
plt.imshow(data)

# Add a color bar to show the scale
plt.colorbar()

# Add labels
plt.title('Heatmap Example')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the heatmap
plt.show()
```
